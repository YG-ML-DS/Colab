```python
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.legend(['sin', 'cos'])

plt.subplot(2,1,1)
plt.plot(x, y_sin)
plt.title('sin')
plt.subplot(2,1,2)
plt.plot(x, y_cos)
plt.title('cos')

plt.show()

```

```python
import pandas as pd
url = "https://raw.githubusercontent.com/byuidatascience/data4python4ds/master/data-raw/flights/flights.csv"
flights = pd.read_csv(url)
```

```python
flights.head()
```

```python
flights["time_hour"] = pd.to_datetime(flights["time_hour"], format="%Y-%m-%dT%H:%M:%SZ")
flights["time_hour"].info()
# revise format of date time
```

```python
( flights.query("dest == 'IAH'").groupby(["year","month"])[["arr_delay"]].mean())
# pivot table work: .query --> filtering  rows
# can group multiple criteria into one function
flights.sort_values(["year", "month", "day", "dep_time"])
# for summary statistics (more than one, use 'agg')
year_delay = (
    flights.groupby(["year"]).agg(
        mean_delay =("dep_delay", "mean"),
        count_flights = ("dep_delay", "count")
    )
)
year_delay
```

```python
(
    flights.assign(
        gain = lambda x: x["dep_delay"] - x["arr_delay"]
    )
)
# working with column, generate a new var based on existed var
# lamda is a calculation function: perform the operation on every row.
```
